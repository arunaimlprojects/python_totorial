### **Tuple Quiz (20 Questions)**
#### **Basic Questions**
1. What is a tuple in Python?  
   a) A mutable data structure  
   b) An immutable data structure  
   c) A dictionary  
   d) A function  

2. How do you create an empty tuple?  
   a) `t = []`  
   b) `t = {}`  
   c) `t = ()`  
   d) `t = None`  

3. Which of the following is a correct way to create a tuple with a single element?  
   a) `t = (10,)`  
   b) `t = (10)`  
   c) `t = [10]`  
   d) `t = {10}`  

4. What is the main difference between a tuple and a list?  
   a) Tuples are mutable, lists are immutable  
   b) Tuples are immutable, lists are mutable  
   c) Tuples store only numbers  
   d) Lists take less memory than tuples  

5. Which function is used to find the length of a tuple?  
   a) `size()`  
   b) `count()`  
   c) `length()`  
   d) `len()`  

#### **Indexing & Slicing**
6. How do you access the first element of a tuple `t = (10, 20, 30, 40)`?  
   a) `t[1]`  
   b) `t[0]`  
   c) `t[-1]`  
   d) `t.first()`  

7. How do you access the last element of a tuple?  
   a) `t[-1]`  
   b) `t[0]`  
   c) `t[1]`  
   d) `t.last()`  

8. What will be the output of `t = (10, 20, 30, 40); print(t[1:3])`?  
   a) `(20, 30, 40)`  
   b) `(20, 30)`  
   c) `(10, 20, 30)`  
   d) `(30, 40)`  

9. What happens if you try to change an element in a tuple?  
   a) It gets updated  
   b) It raises an error  
   c) It creates a copy of the tuple  
   d) It adds a new element  

10. How do you find the index of an element in a tuple?  
    a) `t.find(value)`  
    b) `t.index(value)`  
    c) `t.locate(value)`  
    d) `t.search(value)`  

#### **Tuple Operations**
11. How do you concatenate two tuples `t1 = (1, 2)` and `t2 = (3, 4)`?  
    a) `t1 + t2`  
    b) `t1 * t2`  
    c) `t1.concat(t2)`  
    d) `t1.append(t2)`  

12. What will be the output of `('a', 'b') * 3`?  
    a) `('a', 'b', 'a', 'b', 'a', 'b')`  
    b) `('a', 'b', 'a', 'b')`  
    c) `('aaa', 'bbb')`  
    d) `('a', 'a', 'a', 'b', 'b', 'b')`  

13. How do you check if an element exists in a tuple `t = (10, 20, 30)`?  
    a) `exists(20, t)`  
    b) `t.contains(20)`  
    c) `20 in t`  
    d) `t.has(20)`  

14. What will be the output of `t = (1, 2, 3, 4, 5); print(t[-3:])`?  
    a) `(1, 2, 3)`  
    b) `(3, 4, 5)`  
    c) `(2, 3, 4)`  
    d) `(5, 4, 3)`  

15. What is the result of `t = (10, 20, 30, 40); print(min(t))`?  
    a) `10`  
    b) `20`  
    c) `30`  
    d) `40`  

#### **Advanced Questions**
16. Can a tuple contain a list as an element?  
    a) Yes  
    b) No  

17. What is the output of `t = (1, 2, (3, 4)); print(t[2][1])`?  
    a) `3`  
    b) `4`  
    c) `2`  
    d) `(3, 4)`  

18. What happens when you apply `sorted()` to a tuple?  
    a) It modifies the tuple  
    b) It returns a sorted list  
    c) It returns a sorted tuple  
    d) It throws an error  

19. How do you count occurrences of an element in a tuple?  
    a) `t.count(value)`  
    b) `t.index(value)`  
    c) `t.find(value)`  
    d) `t.search(value)`  

20. What is the output of `t = (10, 20, 30); print(type(t))`?  
    a) `<class 'tuple'>`  
    b) `<class 'list'>`  
    c) `<class 'dict'>`  
    d) `<class 'set'>`
