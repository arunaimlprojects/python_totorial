### **1. Arithmetic Operators (गणितीय ऑपरेटर)**
यह ऑपरेटर्स गणितीय ऑपरेशंस के लिए उपयोग होते हैं।  
**ऑपरेटर्स:** `+`, `-`, `*`, `/`, `//`, `%`, `**`

| ऑपरेटर | उपयोग | विवरण                  |
|--------|-------|------------------------|
| `+`    | `a + b` | जोड़ना                 |
| `-`    | `a - b` | घटाना                 |
| `*`    | `a * b` | गुणा                   |
| `/`    | `a / b` | भाग (decimal में)       |
| `//`   | `a // b`| भाग (integer में)       |
| `%`    | `a % b` | शेष निकालना           |
| `**`   | `a ** b`| घातांक (exponentiation)|

**उदाहरण Code:**
```python
a = 10
b = 3

print("Addition:", a + b)  # 10 + 3 = 13
print("Subtraction:", a - b)  # 10 - 3 = 7
print("Multiplication:", a * b)  # 10 * 3 = 30
print("Division:", a / b)  # 10 / 3 = 3.333...
print("Floor Division:", a // b)  # 10 // 3 = 3
print("Modulus:", a % b)  # 10 % 3 = 1
print("Exponent:", a ** b)  # 10**3 = 1000
```

---

### **2. Comparison Operators (तुलना ऑपरेटर)**
ये ऑपरेटर्स दो मानों की तुलना करते हैं।  
**ऑपरेटर्स:** `==`, `!=`, `>`, `<`, `>=`, `<=`

| ऑपरेटर | उपयोग    | विवरण                |
|--------|----------|----------------------|
| `==`   | `a == b` | दोनों मान समान हैं? |
| `!=`   | `a != b` | दोनों मान भिन्न हैं? |
| `>`    | `a > b`  | बड़ा?               |
| `<`    | `a < b`  | छोटा?              |
| `>=`   | `a >= b` | बड़ा या बराबर?       |
| `<=`   | `a <= b` | छोटा या बराबर?       |

**उदाहरण Code:**
```python
x = 5
y = 8

print(x == y)  # False
print(x != y)  # True
print(x > y)   # False
print(x < y)   # True
print(x >= y)  # False
print(x <= y)  # True
```

---

### **3. Logical Operators (तर्कसंगत ऑपरेटर)**
ये Boolean conditions पर काम करते हैं।  
**ऑपरेटर्स:** `and`, `or`, `not`

| ऑपरेटर | विवरण                                              |
|--------|----------------------------------------------------|
| `and`  | सभी conditions True हों तो परिणाम True होगा।       |
| `or`   | कोई भी एक condition True हो तो परिणाम True होगा।   |
| `not`  | स्थिति को उल्टा कर देता है (True -> False, False -> True)। |

**उदाहरण Code:**
```python
a = True
b = False

print(a and b)  # False
print(a or b)   # True
print(not a)    # False
print(not b)    # True
```

---

### **4. Assignment Operators (असाइनमेंट ऑपरेटर)**
यह मान असाइन (दिए) करने के लिए उपयोग होते हैं।  
**ऑपरेटर्स:** `=`, `+=`, `-=`, `*=`, `/=`, `%=`, `**=`, `//=`

**उदाहरण Code:**
```python
x = 10  # Assign
print("x =", x)

x += 5  # x = x + 5
print("x after += 5:", x)

x -= 3  # x = x - 3
print("x after -= 3:", x)

x *= 2  # x = x * 2
print("x after *= 2:", x)

x /= 4  # x = x / 4
print("x after /= 4:", x)

x %= 2  # x = x % 2
print("x after %= 2:", x)

x = 2
x **= 3  # x = x ** 3
print("x after **= 3:", x)
```

---

### **5. Bitwise Operators (बिटवाइज ऑपरेटर)**
ये bit level पर काम करते हैं।  
**ऑपरेटर्स:** `&`, `|`, `^`, `~`, `<<`, `>>`

| ऑपरेटर | विवरण                     |
|--------|---------------------------|
| `&`    | AND                     |
| `|`    | OR                      |
| `^`    | XOR                     |
| `~`    | NOT                     |
| `<<`   | बायें शिफ्ट (Left Shift) |
| `>>`   | दायें शिफ्ट (Right Shift)|

**उदाहरण Code:**
```python
x = 5  # Binary: 101
y = 3  # Binary: 011

print("x & y =", x & y)  # 101 & 011 = 001 = 1
print("x | y =", x | y)  # 101 | 011 = 111 = 7
print("x ^ y =", x ^ y)  # 101 ^ 011 = 110 = 6
print("~x =", ~x)        # ~101 = -110 = -6
print("x << 1 =", x << 1)  # 101 << 1 = 1010 = 10
print("x >> 1 =", x >> 1)  # 101 >> 1 = 10 = 2
```

---

### **6. Membership Operators (सदस्यता ऑपरेटर)**
यह जाँचते हैं कि कोई मान sequence में है या नहीं।  
**ऑपरेटर्स:** `in`, `not in`

**उदाहरण Code:**
```python
lst = [1, 2, 3, 4]

print(2 in lst)  # True
print(5 in lst)  # False
print(5 not in lst)  # True
```

---

### **7. Identity Operators (पहचान ऑपरेटर)**
यह identity की जाँच के लिए उपयोग होते हैं।  
**ऑपरेटर्स:** `is`, `is not`

**उदाहरण Code:**
```python
a = 10
b = 10
c = [10]

print(a is b)      # True (समान object)
print(a is not c)  # True (अलग object)
```

---

### **8. Special Operators:**

#### **Ternary Operator (तीन-आयामी ऑपरेटर)**
एक line में condition लिखने के लिए उपयोग होता है।  
**Syntax:** `value_if_true if condition else value_if_false`

**उदाहरण Code:**
```python
a = 5
b = 10

result = "a is greater" if a > b else "b is greater"
print(result)  # Output: b is greater
```

4. Assignment Operators (असाइनमेंट ऑपरेटर)

 =, +=, -=, *=, /=, %=, **=, //=

````python
x = 10  # Assign
print("x =", x)

x += 5  # x = x + 5
print("x after += 5:", x)

x -= 3  # x = x - 3
print("x after -= 3:", x)

x *= 2  # x = x * 2
print("x after *= 2:", x)

x /= 4  # x = x / 4
print("x after /= 4:", x)

x %= 2  # x = x % 2
print("x after %= 2:", x)

x = 2
x **= 3  # x = x ** 3
print("x after **= 3:", x)
````

5. Bitwise Operators (बिटवाइज ऑपरेटर)

&, |, ^, ~, <<, >>

````python
x = 5  # Binary: 101
y = 3  # Binary: 011

print("x & y =", x & y)  # 101 & 011 = 001 = 1
print("x | y =", x | y)  # 101 | 011 = 111 = 7
print("x ^ y =", x ^ y)  # 101 ^ 011 = 110 = 6
print("~x =", ~x)        # ~101 = -110 = -6
print("x << 1 =", x << 1)  # 101 << 1 = 1010 = 10
print("x >> 1 =", x >> 1)  # 101 >> 1 = 10 = 2
````
