````python
# 1. Min of 3 Numbers
a = int(input('first num: '))
b = int(input('second num: '))
c = int(input('third num: '))

if a < b and a < c:
    print('smallest is', a)
elif b < c:
    print('smallest is', b)
else:
    print('smallest is', c)
````

````python
# 2. Menu-Driven Calculator
menu = input("""
Hi! how can I help you.
1. Enter 1 for pin change
2. Enter 2 for balance check
3. Enter 3 for withdrawl
4. Enter 4 for exit
""")

if menu == '1':
  print('pin change')
elif menu == '2':
  print('balance')
else:
  print('exit')
````

````python
Here are some simple questions and answers using if-elif-else:

---

### **Question 1:**  
Given a number. If it is greater than 0, print "Positive". If it is less than 0, print "Negative". If it is equal to 0, print "Zero".

**Answer:**  
```python
num = int(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

---

### **Question 2:**  
Given a person's age. If the age is 18 or greater, print "Eligible to vote". Otherwise, print "Not eligible to vote".

**Answer:**  
```python
age = int(input("Enter your age: "))

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")
```

---

### **Question 3:**  
Given a number. If it is divisible by 2, print "Even". Otherwise, print "Odd".

**Answer:**  
```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

---
````

````python
Here are the answers to the 10 simple if-else questions:

---

### **Question 1 Answer:**  
You are given a number. If it is even, print "Even", otherwise print "Odd".
```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

---

### **Question 2 Answer:**  
You are given a number. If it is positive, print "Positive", otherwise print "Negative".
```python
num = int(input("Enter a number: "))

if num > 0:
    print("Positive")
else:
    print("Negative")
```

---

### **Question 3 Answer:**  
Check if the entered year is a leap year. A leap year is divisible by 4 but not divisible by 100 unless also divisible by 400.
```python
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

---

### **Question 4 Answer:**  
Given a number, check if it is divisible by both 5 and 10.
```python
num = int(input("Enter a number: "))

if num % 5 == 0 and num % 10 == 0:
    print("Divisible by both 5 and 10")
else:
    print("Not divisible by both 5 and 10")
```

---

### **Question 5 Answer:**  
If the given age is 18 or greater, print "Adult". Otherwise, print "Minor".
```python
age = int(input("Enter your age: "))

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

---

### **Question 6 Answer:**  
You have a number. If it is greater than 50, print "Greater", otherwise print "Smaller".
```python
num = int(input("Enter a number: "))

if num > 50:
    print("Greater")
else:
    print("Smaller")
```

---

### **Question 7 Answer:**  
Given a grade, if the grade is A, print "Excellent", if B, print "Good", and if C or below, print "Needs Improvement".
```python
grade = input("Enter your grade: ")

if grade == 'A':
    print("Excellent")
elif grade == 'B':
    print("Good")
else:
    print("Needs Improvement")
```

---

### **Question 8 Answer:**  
You are given a number. If the number is divisible by 3, print "Fizz", if divisible by 5, print "Buzz". If divisible by both, print "FizzBuzz".
```python
num = int(input("Enter a number: "))

if num % 3 == 0 and num % 5 == 0:
    print("FizzBuzz")
elif num % 3 == 0:
    print("Fizz")
elif num % 5 == 0:
    print("Buzz")
else:
    print("Not divisible by 3 or 5")
```

---

### **Question 9 Answer:**  
Check if the entered number is in the range of 1 to 100 (both inclusive).
```python
num = int(input("Enter a number: "))

if 1 <= num <= 100:
    print("Number is in the range of 1 to 100")
else:
    print("Number is out of range")
```

---

### **Question 10 Answer:**  
You are given a string. If its length is greater than 5, print "Long", otherwise print "Short".
```python
string = input("Enter a string: ")

if len(string) > 5:
    print("Long")
else:
    print("Short")
```

---

These are the solutions for the questions provided. Let me know if you need further clarification!
````
