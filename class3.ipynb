यहाँ पर सिर्फ **if-else** से जुड़े सवाल दिए गए हैं। आप इन सवालों को हल करने की कोशिश करें:

---

### 1. **नंबर की पहचान:**
अगर कोई नंबर 0 है तो "Zero" प्रिंट करें, अन्यथा "Non-Zero" प्रिंट करें।

---

### 2. **संख्या को Odd या Even में वर्गीकृत करें:**
किसी भी संख्या को Odd (विषम) या Even (सम) के रूप में चेक करें।

---

### 3. **पॉजिटिव या नेगेटिव:**
अगर नंबर पॉजिटिव है तो "Positive" प्रिंट करें, अन्यथा "Negative" प्रिंट करें।

---

### 4. **उम्र चेक करें:**
अगर कोई व्यक्ति 18 साल का है या उससे ज्यादा तो "Eligible for Voting" प्रिंट करें, अन्यथा "Not Eligible" प्रिंट करें।

---

### 5. **ग्रेड असाइन करें:**
अगर छात्र के नंबर 90 या उससे अधिक हैं, तो "A" दें; 75-89 के बीच "B", अन्यथा "C"।

---

### 6. **साल Leap Year है या नहीं:**
चेक करें कि एक दिया गया वर्ष Leap Year है या नहीं।

---

### 7. **दो नंबर की तुलना:**
अगर पहला नंबर दूसरे से बड़ा है तो "First is Greater" प्रिंट करें, अन्यथा "Second is Greater"।

---

### 8. **सप्ताहांत चेक करें:**
अगर दिन "Saturday" या "Sunday" है तो "Weekend" प्रिंट करें, अन्यथा "Weekday"।

---

### 9. **पास या फेल चेक करें:**
अगर स्टूडेंट के अंक 40 से अधिक हैं तो "Pass" प्रिंट करें, अन्यथा "Fail"।

---

### 10. **समय के आधार पर संदेश:**
अगर समय 12 बजे से पहले है तो "Good Morning" प्रिंट करें, अन्यथा "Good Afternoon"।

---

इन सवालों को हल करने के लिए अपने कोड लिखने की कोशिश करें। किसी समस्या में मदद चाहिए तो बताइए! 😊

Here are **if-else** based questions in English for practice. Try solving them:

---

### 1. **Identify a Number:**
Check if a given number is `0`. If yes, print "Zero," otherwise print "Non-Zero."

---

### 2. **Odd or Even:**
Check whether a given number is odd or even.

---

### 3. **Positive or Negative:**
Check if a number is positive. If yes, print "Positive," otherwise print "Negative."

---

### 4. **Voting Eligibility:**
If a person's age is 18 or more, print "Eligible for Voting"; otherwise, print "Not Eligible."

---

### 5. **Assign a Grade:**
If a student's marks are 90 or above, assign "A"; for marks between 75 and 89, assign "B"; otherwise, assign "C."

---

### 6. **Leap Year or Not:**
Check whether a given year is a leap year.

---

### 7. **Compare Two Numbers:**
If the first number is greater than the second, print "First is Greater"; otherwise, print "Second is Greater."

---

### 8. **Check for Weekend:**
If the day is "Saturday" or "Sunday," print "Weekend"; otherwise, print "Weekday."

---

### 9. **Pass or Fail:**
If a student's marks are greater than 40, print "Pass"; otherwise, print "Fail."

---

### 10. **Time-Based Message:**
If the current time is before 12:00 PM, print "Good Morning"; otherwise, print "Good Afternoon."

---
